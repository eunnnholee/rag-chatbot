```python
import numpy as np
import pandas as pd
from gensim import corpora, models
from gensim.models.coherencemodel import CoherenceModel
```

```python
df = pd.read_csv('konlpy이후.csv', encoding="utf-8", index_col=0)
df
```

```python
stop_words = "것 저 수 분 데 중 전 일 때 이 안녕 육아휴직 육휴 육아 휴직 출산휴가" #추가적으로 명사 중 쓸모없는 단어 제거
stop_words=stop_words.split(' ')
stop_words
def remove_stopwords(text, stopwords): # stopwords를 제거하는 함수.
    return " ".join([word for word in str(text).split() if word not in stopwords])

df['nouns'] = df['nouns'].apply(lambda post: remove_stopwords(post, stop_words))
```

```python
# BOW 벡터 구하기

posts = [x.split(' ') for x in df['nouns']]
id2word = corpora.Dictionary(posts) # 각 문서에서 나타나는 모든 단어를 추출하여 단어 사전을 만듬. 단어 사전은 각 단어에 대한 고유 ID할당.
# ID를 이용하여 각 문서를 벡터로 표현할 수 있음
corpus_tf = [id2word.doc2bow(text) for text in posts]
# posts라는 문서 집합의 각 문서에 대해 각 문서에서 나타나는 단어의 빈도를 계산하여 BOW 벡터를 만듬
# BOW 벡터 : 각 단어의 빈도를 나타냄 ex(0,1) : ID가 0인 단어의 빈도는 1이다.
print(corpus_tf[0]) # 첫 번째 문서의 BOW 벡터를 출력
```

```python
# TF-IDF 값이 5 이상인 단어를 리스트로 출력

tfidf = models.TfidfModel(corpus_tf)
corpus_tfidf = tfidf[corpus_tf]
# print(corpus_tfidf[0])

# 단어별 TF-IDF 총합 계산
word_tfidf_sums = {}

for doc in corpus_tfidf:
    for word_id, tfidf_value in doc:
        word = id2word[word_id]
        if word in word_tfidf_sums:
            word_tfidf_sums[word] += tfidf_value
        else:
            word_tfidf_sums[word] = tfidf_value

# TF-IDF 총합을 기준으로 내림차순 정렬
sorted_word_tfidf_sums = sorted(word_tfidf_sums.items(), key=lambda x: x[1], reverse=True)

# TF-IDF 값이 5 이상인 단어 출력
ex = []
for word, tfidf_sum in sorted_word_tfidf_sums:
    if tfidf_sum >= 5:
        ex.append(word)
    else:
        continue
print(ex)

# 의미 없는 단어 제거
x = "다음날 연락 15일 질문드 몸 얘 바 날 그 2일 축 10일 이거 동안 29일 육 여기 앞 되는거 둘다 인정 예정인데요 곳 150 3일 5일 어느정도 12일 이럴경우 21일 이런경우 4일 진짜 걸 건가요 20일 개 19일 14일 육아휴직이요 4 정확 대 2023 봐주세용 관련해서 몇 체크 줄 정도 위 듯 라 누구 끝 편 육아휴직1년 반 어제 의견 예 있는거 다음주 25 부 25일 23일 17일 5 13일 100 일요일 금요일 한 안쓰 뭔가 45 18일 이런식 23 31 궁금 1월 2월 3월 4월 5월 6월 7월 8월 9월 10월 11월 12월 1일 다들 근데 1 들 감사 애 부탁드 부터 관련 오늘 2 지 무 후 내 어마 질문이요 나 계 저희 되 3 문의드 이번 맘님들 있을까요 그때 리 자 첫 문 적 육아휴직중이 안 거"
remove_list = x.split(' ')
for remove_word in remove_list:
    try:
        ex.remove(remove_word)
    except:
        continue
print(ex)
print(len(ex))

# 결과 출력
# for word, tfidf_sum in sorted_word_tfidf_sums:
#     print(f"Word: {word}, TF-IDF Sum: {tfidf_sum}")
```

```python
texts = df['nouns']
# 단어 쌍의 동시 출현 빈도를 저장할 딕셔너리
cooccur = {}

for text in texts:
    # 텍스트를 단어로 분리
    text = str(text)     # text변수 안의 문자를 문자열로 변환 / 숫자나, 불리안을 문자로
    words = text.split() # 공백을 기준으로 분리
    words = set(words)   # 중복 단어 제거
    words = list(words)  
    
    # 단어 쌍의 동시 출현 빈도 
    for i in range(len(words)):
        for j in range(i+1, len(words)):
            word1 = words[i]
            word2 = words[j]
            if (word1, word2) in cooccur:
                cooccur[(word1, word2)] += 1
            elif (word2, word1) in cooccur:
                cooccur[(word2, word1)] += 1
            else:
                cooccur[(word1, word2)] = 1
                
# 동시출현빈도 리스트화
cooccur_list = list(cooccur.items())

# 선정된 키워드로 구성된 edge 추출
final_edgelist = []                
for pair, weight in cooccur_list:
    if pair[0] in ex:
        if pair[1] in ex:
            final_edgelist.append(((pair[0], pair[1]), weight))

print(final_edgelist)
print(len(final_edgelist))
                
final_nodelist = ex
```

```python
# SNA를 위한 final_edgelist 데이터프레임화 후 저장

df1 = pd.DataFrame(final_edgelist, columns=['Word', 'Weight'])


series_split1 = df1["Word"].apply(lambda x: x[0])
series_split2 = df1["Word"].apply(lambda x: x[1])
df1["Weight"]

SNA_data = pd.DataFrame({
    "Source": series_split1,
    "Target": series_split2,
    "Weight": df1["Weight"]
})

# SNA_data.to_csv("SNA연습.csv", encoding="cp949")
```
