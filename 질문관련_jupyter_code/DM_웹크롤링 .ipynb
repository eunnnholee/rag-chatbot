```python
from selenium import webdriver
import chromedriver_autoinstaller
import time
import pandas as pd
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from selenium.webdriver.common.alert import Alert
# import openpyxl

# 크롬버전 확인
chrome_ver = chromedriver_autoinstaller.get_chrome_version().split('.')[0]
print(chrome_ver)
```

```python
try:
    crawler = webdriver.Chrome(f'./{chrome_ver}/chromedriver.exe')
except:
    chromedriver_autoinstaller.install(True)
    crawler = webdriver.Chrome(f'./{chrome_ver}/chromedriver.exe')
    
crawler.implicitly_wait(10)


# 필요한 크롬확장자 설치 (확장자 설치시 확인버튼 사용자가 눌러야함.)
url2 = 'https://chrome.google.com/webstore/detail/%EB%84%A4%EC%9D%B4%EB%B2%84-%EC%B9%B4%ED%8E%98-%ED%8C%A8%EC%8A%A4/gipgjcnhbklggnannochejcaieghkmcn?hl=ko'
crawler.get(url2)
crawler.find_element('xpath', '/html/body/div[3]/div[2]/div/div/div[2]/div[2]/div/div/div/div').click()
time.sleep(5)
url3 = 'https://chrome.google.com/webstore/detail/%EB%84%A4%EC%9D%B4%EB%B2%84-%EC%B9%B4%ED%8E%98-%EC%83%88%EB%A1%9C%EA%B3%A0%EC%B9%A8/jlebnhjlcighfebijokkgijldbfiameg?hl=ko'
crawler.get(url3)
crawler.find_element('xpath', '/html/body/div[3]/div[2]/div/div/div[2]/div[2]/div/div').click()
time.sleep(5)
#확인 눌러야함
url = 'https://cafe.naver.com/imsanbu'
crawler.get(url)
```

```python
# 주어진 게시판(xpath)로 가서 크롤링하기
title = []
detail = []
def navercafe(xpath,n): # n은 총 페이지 수
    carrier_mom = crawler.find_element('xpath', xpath).get_attribute("href") # 게시판 링크
    crawler.get(carrier_mom)
    crawler.switch_to.frame("cafe_main")
    engine = crawler.find_element('xpath', '//*[@id="query"]')
    engine.click()
    engine.send_keys('육아휴직') # 검색엔진에 입력하고 싶은 키워드 쓰기
    engine.send_keys(Keys.ENTER) # 검색엔진의 엔터를 눌러줘서 검색을 가능하게 해준다.
    for i in range(1, n):
        print('----' + str(i) + ' 번째 페이지 -----')
        for a in range(1, 16): # 게시판 페이지 수 
            b = crawler.find_element('xpath',f'//*[@id="main-area"]/div[5]/table/tbody/tr[{a}]/td[1]/div[2]/div/a').get_attribute("href")
            crawler.get(b)
            crawler.switch_to.frame("cafe_main")
            b = crawler.find_element('xpath','//*[@id="app"]/div/div/div[2]/div[1]/div[1]/div/h3').text
            print("제목 : "+b)
            title.append(b)
            try:
                b = crawler.find_element('xpath','//*[@id="app"]/div/div/div[2]/div[2]/div[1]/div[2]/div[1]').text
            except:
                b = crawler.find_element('xpath','//*[@id="app"]/div/div/div[2]/div[2]/div[1]/div/div[1]').text
            print("내용 : "+b, sep='\n')
            detail.append(b)
            time.sleep(1)
            crawler.back()
            time.sleep(1)
            crawler.switch_to.frame("cafe_main")
            
        if i < n-1:
            if i <= 10:
                if i%10 != 0:
                    crawler.find_element('xpath',f'//*[@id="main-area"]/div[7]/a[{i%10}+1]').click()
                else : 
                    crawler.find_element('xpath','//*[@id="main-area"]/div[7]/a[11]/span').click()
            else:
                if i%10 != 0:
                    crawler.find_element('xpath',f'//*[@id="main-area"]/div[7]/a[{i%10}+2]').click()
                else : 
                    crawler.find_element('xpath','//*[@id="main-area"]/div[7]/a[12]/span').click()  
    crawler.get(url)

# 직장맘 수다방 100개 22/10/05  '//*[@id="menuLink135"]'
# 임신 중 질문방 60개 22/02/04  '//*[@id="menuLink392"]'
# 육아 질문방 15개 22/03/08    '//*[@id="menuLink126"]'
# 자유 수다방 100개 22/04/20   '//*[@id="menuLink179"]'
```

```python
navercafe('//*[@id="menuLink135"]',101)
navercafe('//*[@id="menuLink392"]',61)
navercafe('//*[@id="menuLink126"]',16)
navercafe('//*[@id="menuLink179"]',101)

```

```python
# 크롤링 결과 csv파일로 저장
import pandas as pd
df = pd.DataFrame({"Title":title, "Detail":detail})
df["new"] = df["Title"] + " " + df["Detail"]
df.to_csv("변환파일.csv", encoding="cpc949", index_col=0)
```
